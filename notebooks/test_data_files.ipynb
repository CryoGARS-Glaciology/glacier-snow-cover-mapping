```python
import os
import glob
import xarray as xr
import rioxarray as rxr
import pandas as pd
import geopandas as gpd
```

## Define path to data and site names

```python
data_path = '/Users/raineyaberle/Google Drive/My Drive/Research/PhD/write-ups/CH1_snow_cover_mapping_methods_manuscript/Aberle_et_al_dataset_submission'
site_names = ['Gulkana', 'Wolverine', 'LemonCreek', 'Sperry', 'SouthCascade']
site_names
```

## Open classified image netCDF files, check for correct CRS and attributes

```python
for site_name in site_names:
    print(site_name)
    
    im_classified_path = os.path.join(data_path, site_name, 'imagery', 'classified')
    im_classified_fns = glob.glob(os.path.join(im_classified_path, '*.nc'))

    for im_classified_fn in im_classified_fns:
        im = rxr.open_rasterio(im_classified_fn)
        crs = im.rio.crs.to_epsg()
        print(crs)
    
    print(' ')
```

## Open snow cover statistics CSVs, check for correct CRS and attributes

```python
site_names = ['Wolverine']
for site_name in site_names:
    print(site_name)
    
    snow_cover_stats_path = os.path.join(data_path, site_name, 'imagery', 'snowlines')
    stats_fns = glob.glob(os.path.join(snow_cover_stats_path, '*.csv'))

    # load and concatenate snow cover stats files
    stats_df = pd.DataFrame()
    for stats_fn in stats_fns:
        try:
            stats = pd.read_csv(stats_fn)
            stats_df = pd.concat([stats_df, stats])
        except:
            print(stats_fn + ': Error opening')

    # check that columns are the same for all files
    
    print(' ')
```
